```python
import numpy as np
import FortranDriver
md = FortranDriver.MDInterface()
```

```python
#Setting parameters
md._npart = 500
print(f"Num. particles = {md.npart}")
md._kf = 1.0
print(f"WCA epsilon = {md.eps}") #Is it kf or eps1/eps2 that determines the value of epsilon?
md._temp = 5.0
md._tr = 5.0
print(f"Temperature? = {md._temp}")
print(f"Temperature?? = {md._tr}") # Which one is temperature? 
md._fe0 = 1.0
print(f"Field strength = {md._fe0}")
md._drf = 1.0
print(f"Reduced density = {md._drf}")
# I'm actually not sure that these are really changing the parameters - my RDF and 2D plot looks the same no matter what parameters I set
# Also something is perhaps going wrong in changing parameters in the Fortran driver as well?
# i.e. If I change fe0 in the driver, save it, reload md and do print "print(f"Field strength = {md._fe0}")", it doesn't reflect the change I've made 
```

```python
# Initial configuration (3D)
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(md.x, md.y, md.z, 'o')
```

```python
# Initial configuration (2D)
plt.plot(md.x, md.y, 'o')
```

```python
# Graphing the RDF (pre-simulation)
rdf_dict = md.compute_rdf()
r_array = np.array(rdf_dict['r'])
rdf_array = np.array(rdf_dict['rdf'])
plt.plot(r_array,rdf_array,"-k")
plt.title("g(r) Radial Distribution Function (pre-simulation)")
```

```python
# Running simulation and final configuration (2D)
md.run(5000)
plt.plot(md.x, md.y, 'o')
```
